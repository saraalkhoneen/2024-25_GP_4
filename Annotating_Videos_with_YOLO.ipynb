# **Image annotation using YOLOv11.**

```python
#Download ultralytics
%pip install ultralytics
import ultralytics
```

To control the number of images extracted from each video, I created a separate project in Roboflow to split every video into a specific number of frames. Then, I uploaded all the frames into one dataset (training set) for faster access.

```python
#1 Access the dataset
!pip install roboflow
from roboflow import Roboflow

from google.colab import userdata

rf = Roboflow(api_key=userdata.get('roboflowApiKey'))
project = rf.workspace("first-space-wzw87").project("yolo-frames")
version = project.version(2)
dataset = version.download("yolov11")
```

```python
# 2 Load a model
from ultralytics import YOLO

model = YOLO("yolo11n.pt")  # pretrained YOLO11n model

# 3 Start annotating the images with model.predict
model.predict(source='/content/YOLO-Frames---2/train/images', save= True, show_labels= False, save_frames= True, save_txt= True, show_boxes= False)
```

After annotating the frames using YOLO, I am saving both the frames and their corresponding text files to upload them back to Roboflow, to fix any mistakes happened during the annotation process.
